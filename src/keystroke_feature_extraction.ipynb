## Keystroke Feature Extraction

```python
from dataman.audio_processing import *

from librosa.feature import mfcc, melspectrogram
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from scipy.signal import spectrogram
```

```python
visualize_keystrokes('/env/datasets/keystrokes/space/space-x.wav')
```

```python
data, labels_int, labels_str = load_keystroke_data()
```

```python
spacebars = data[labels_str == 'space']
sample = spacebars[0]
```

```python
plt.plot(sample)
plt.show()
```

```python
sample_nonoise = remove_random_noise(sample, )
plt.plot(sample_nonoise)
plt.show()
```

```python
sample_mfcc = mfcc(sample.astype(float))
plt.plot(sample_mfcc)
plt.show()
```

```python
sample_melspec = melspectrogram(sample.astype(float))
plt.plot(sample_melspec)
plt.show()
```
